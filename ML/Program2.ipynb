```python
#perform arithmetic array operation:add,sub,mul,div,exponentiation and modulus using statistics,numpy,scipy,math.
import numpy as np
data = np.array([1, 2, 3, 4, 5])
```

```python
import numpy as np
#Element-wise arithmetic, aggregates, cumulative operations, sqrt/log/exp

# Element-wise arithmetic
print("Original data:", data)
print("Add 5:", data + 5)
print("Subtract 2:", data - 2)
print("Multiply by 3:", data * 3)
print("Divide by 2:", data / 2)
print("Power 2:", data ** 2)
print("Modulus 2:", data % 2)

# Aggregate functions
print("Sum:", np.sum(data))
print("Product:", np.prod(data))
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Variance:", np.var(data))
print("Minimum:", np.min(data))
print("Maximum:", np.max(data))

# Other useful functions
print("Cumulative sum:", np.cumsum(data))
print("Cumulative product:", np.cumprod(data))
print("Square root:", np.sqrt(data))
print("Exponential:", np.exp(data))
print("Logarithm:", np.log(data))
```

```python
#scipy package
#Advanced statistics, integration, optimization, correlation, trimmed mean
from scipy import stats
from scipy import integrate
from scipy import optimize

# Statistics
print("Mean:", stats.tmean(data))
print("Trimmed mean (10% cut):", stats.tmean(data, limits=(1, 5)))
#print("Median:", stats.tmedian(data))----removed in modern scipy
print(median = stats.scoreatpercentile(data, 50))
print("Variance:", stats.tvar(data))
print("Standard Deviation:", stats.tstd(data))
print("Mode:", stats.mode(data))

# Integration example (area under x^2 from 0 to 5)
f = lambda x: x**2
area, error = integrate.quad(f, 0, 5)
print("Integral of x^2 from 0 to 5:", area)

# Optimization example (find minimum of f(x) = (x-3)^2)
g = lambda x: (x - 3)**2
result = optimize.minimize(g, x0=0)
print("Minimum of (x-3)^2:", result.x[0])

# Correlation example
y = np.array([2, 4, 6, 8, 10])
correlation = stats.pearsonr(data, y)
print("Pearson correlation with [2,4,6,8,10]:", correlation[0])
```

```python
#statistics package
#Mean, median, mode, variance, stdev, harmonic/geometric mean
import statistics
data_list = list(data)
data1 = list(map(int,data))
print("Mean:", statistics.mean(data_list))
print("Median:", statistics.median(data_list))
print("Mode:", statistics.mode(data_list))
print("Variance:", statistics.variance(data_list))
print("Standard Deviation:", statistics.stdev(data1))
print("Harmonic mean:", statistics.harmonic_mean(data_list))
print("Geometric mean:", statistics.geometric_mean(data_list))
print("Sum:", sum(data_list))
print("Min:", min(data_list))
print("Max:", max(data_list))
```

```python
#math package
#Scalar operations, factorial, sqrt, log, trig functions, list comprehensions
import math

sqrt_values = [math.sqrt(x) for x in data]
factorials = [math.factorial(x) for x in data]
log_values = [math.log(x) for x in data]
sin_values = [math.sin(x) for x in data]
cos_values = [math.cos(x) for x in data]

print("Square roots:", sqrt_values)
print("Factorials:", factorials)
print("Natural log:", log_values)
print("Sine values:", sin_values)
print("Cosine values:", cos_values)

# Single value math examples
total = sum(data)
print("Log of total sum:", math.log(total))
print("Exponential of total sum:", math.exp(total))
print("Square root of total sum:", math.sqrt(total))
```
