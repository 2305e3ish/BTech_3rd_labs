```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
data = pd.read_csv("salary_data.csv")   # columns: Experience, Salary

X = data[['YearsExperience']]
y = data['Salary']

# Create model
model = LinearRegression()
model.fit(X, y)

# Predict
y_pred = model.predict(X)

# Correlation
correlation = data['YearsExperience'].corr(data['Salary'])
print("Correlation between Experience and Salary:", correlation)

# Plot
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Simple Linear Regression")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
data = pd.read_csv("minihomeprices.csv")

# Fill missing bedroom values with median
data.bedrooms = data.bedrooms.fillna(data.bedrooms.median())

X = data[['area', 'bedrooms', 'age']]
y = data['price']

model = LinearRegression()
model.fit(X, y)

# Predict price for new house
new_house = pd.DataFrame([[3000, 3, 40]], 
                         columns=['area', 'bedrooms', 'age'])

predicted_price = model.predict(new_house)

print("Predicted Price:", predicted_price[0])
```
