```python
# Implement and demonstrate the FIND-S algorithm for finding the most specific hypothesis for a given
# dataset.

# ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes']
# ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes']
# ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No']
# ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']

# Dataset: Each row = [Attr1, Attr2, ..., AttrN, Target]
dataset = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']
]

def find_s_algorithm(data):
    # Initialize hypothesis with most specific values
    hypothesis = ['ϕ'] * (len(data[0]) - 1)

    for example in data:
        if example[-1] == 'Yes':  # Only consider positive examples
            for i in range(len(hypothesis)):
                if hypothesis[i] == 'ϕ':
                    hypothesis[i] = example[i]
                elif hypothesis[i] != example[i]:
                    hypothesis[i] = '?'
    return hypothesis

# Run FIND-S
final_hypothesis = find_s_algorithm(dataset)

print("Final Hypothesis:", final_hypothesis)
```
