```python
# Load Iris dataset and convert to a 1-D numeric list (sepal length)
from sklearn import datasets
iris = datasets.load_iris()
# use sepal length (column 0) so existing statistics code works on a list of numbers
data = iris.data[:, 0].tolist()
print('Loaded iris, n=', len(data), ' — first 10:', data[:10])
```

```python
#central tendencies-inbuilt
from statistics import mean,median,mode
print(mean(data))
print(median(data))
print(mode(data))
```

```python
#central tendencies-logic
from collections import Counter
n  = len(data)
mean = sum(data)/n
data = sorted(data)
median = (data[(n-1)//2] + data[n//2]) / 2
mode = max(set(data), key=data.count)
mode1 = Counter(data).most_common(1)[0][0]
print(Counter(data))
print(f"{mean}\n{median}\n{mode}\n{mode1}")
```

```python
#Dispersion-inbuilt
from statistics import *
population_sd = pstdev(data) #same as manual
sample_sd = stdev(data)
population_var = pvariance(data) #same as manual
sample_var = variance(data)
print(f"{population_sd}\n{sample_sd}\n{population_var}\n{sample_var}")
```

```python
#Dispersion manual
from math import sqrt
variance_manual = sum([(x - sum(data)/n)**2 for x in data]) / n
stdev_manual = sqrt(variance_manual)
print(f"{variance_manual}\n{stdev_manual}")
```
