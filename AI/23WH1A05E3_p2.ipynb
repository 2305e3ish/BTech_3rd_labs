```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
#implement a decision tree classifier for a basic classification problem

# what a general code of decision tree code does step by step: 
# Imports modules for loading data, splitting data, modelling, and evaluating. 
# Loads the Iris dataset, a classic beginner classification dataset built-in scikit-learn. 
# Splits the data into training and testing sets (30% for testing).
# Creates a Decision Tree Classifier object. 
# Trains the tree on the training data. 
# Uses the trained model to predict the labels of the test set and prints the accuracy.
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=99
)
# Defining the Decision Tree Model
clf = DecisionTreeClassifier(
    criterion="entropy",
    max_depth=2,
    random_state=1
)
clf.fit(X_train, y_train)
```

```python
# Making predictions
y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(f'Predicted val:  {y_pred}')
print(f'Actual val:     {y_test}')

print(f'classification report:\n{classification_report(y_test,y_pred)}')
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Increase the figure size so the tree is readable
plt.figure(figsize=(20,10))

# Plot the tree
plot_tree(clf, 
          feature_names=data.feature_names,  # Names of the features (mean radius, etc.)
          class_names=data.target_names,    # Names of the classes (malignant, benign)
          filled=True,                      # Colors the nodes based on the majority class
          rounded=True,                     # Rounds the corners of the boxes
          fontsize=10)

plt.show()
```
