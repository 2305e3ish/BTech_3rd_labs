```python
from itertools import permutations

# Distance matrix
graph = [
    [0, 10, 15, 20],
    [5, 0, 9, 10],
    [6, 13, 0, 12],
    [8, 8, 9, 0]
]

n = len(graph)
cities = [1, 2, 3]  # Excluding starting city 0
min_cost = float('inf')

for p in permutations(cities):
    cost = 0
    k = 0  # start from city 0
    for city in p:
        cost += graph[k][city]
        k = city
    cost += graph[k][0]  # return to start
    if cost < min_cost:
        min_cost = cost
        best_path = (0,) + p + (0,)

print("Minimum Cost:", min_cost)
print("Best Path:", best_path)
```
