```python
from collections import deque

# Function to solve the Monkey Banana Problem using Breadth-First Search
def solve_monkey_banana():
    # State representation: (Monkey Position, Box Position, Monkey on Box?, Has Banana?)
    # Room locations: 'Door', 'Window', 'Middle' (Banana is hanging in the Middle)
    start_state = ('Door', 'Window', False, False)
    
    # Queue stores tuples of (current_state, path_of_actions)
    queue = deque([(start_state, [])])
    visited = set([start_state])
    
    while queue:
        current_state, path = queue.popleft()
        m_pos, b_pos, on_box, has_banana = current_state
        
        # Goal Check
        if has_banana:
            print("Goal Reached! Sequence of actions:")
            for step, action in enumerate(path, 1):
                print(f"{step}. {action}")
            return
            
        # Generate possible valid moves
        moves = []
        
        # 1. Walk to the box
        if not on_box and m_pos != b_pos:
            moves.append(( (b_pos, b_pos, False, False), f"Walk to {b_pos}" ))
            
        # 2. Push the box to the Middle (where the banana is)
        if not on_box and m_pos == b_pos and b_pos != 'Middle':
            moves.append(( ('Middle', 'Middle', False, False), "Push box to Middle" ))
            
        # 3. Climb onto the box
        if not on_box and m_pos == b_pos:
            moves.append(( (m_pos, b_pos, True, False), "Climb onto the box" ))
            
        # 4. Grasp the banana
        if on_box and b_pos == 'Middle' and not has_banana:
            moves.append(( (m_pos, b_pos, True, True), "Grasp the banana" ))
            
        # Add valid next states to the queue
        for next_state, action in moves:
            if next_state not in visited:
                visited.add(next_state)
                queue.append((next_state, path + [action]))
                
    print("No solution found.")

# Run the algorithm
solve_monkey_banana()
```
