```python
def water_jug_dfs(cap1, cap2, target):
    visited = set()

    def dfs(x, y, path):
        # If already visited, stop
        if (x, y) in visited:
            return False
        # Mark visited
        visited.add((x, y))
        path.append((x, y))
        # Check goal
        if x == target or y == target:
            print("Solution Found:")
            for step in path:
                print(step)
            return True
        # Generate all possible moves
        # 1. Fill Jug1
        if dfs(cap1, y, path.copy()):
            return True
        # 2. Fill Jug2
        if dfs(x, cap2, path.copy()):
            return True
        # 3. Empty Jug1
        if dfs(0, y, path.copy()):
            return True
        # 4. Empty Jug2
        if dfs(x, 0, path.copy()):
            return True
        # 5. Pour Jug1 → Jug2
        transfer = min(x, cap2 - y)
        if dfs(x - transfer, y + transfer, path.copy()):
            return True
        # 6. Pour Jug2 → Jug1
        transfer = min(y, cap1 - x)

        if dfs(x + transfer, y - transfer, path.copy()):
            return True
        return False

    # Start from (0,0)
    if not dfs(0, 0, []):
        print("No Solution Possible")

# Example: 5L and 3L jugs, target 4L
water_jug_dfs(5, 3, 4)
```
